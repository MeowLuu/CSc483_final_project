```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# NOTE: run pip install sentence-transformers transformers scikit-learn as an admin on terminal before running this

model = SentenceTransformer('bert-base-nli-mean-tokens')

def calculate_similarity(query, dataset):
    query_embedding = model.encode([query])
    similarities = []

    for item in dataset:
        item_embedding = model.encode([item])
        similarity = cosine_similarity(query_embedding, item_embedding)[0][0]
        similarities.append((item, similarity))

    # Sort by similarity in descending order, i.e the most relevant categories at the front
    similarities.sort(key=lambda x: x[1], reverse=True)
    return similarities
```

```python
user_ids = user_data['user_id']
print(user_ids.head(20))
user_ids = user_ids.values.tolist()
```

```python
def get_category(user_query: str):
    results = calculate_similarity(user_query, categories)
    return results[0][0]
```

```python
def preprocessing(input_data: pd.DataFrame):
    encoder = OrdinalEncoder()
    input_data['name'] = encoder.fit_transform(input_data['name'].values.reshape(-1, 1))
    input_data['address'] = encoder.fit_transform(input_data['address'].values.reshape(-1, 1))
    input_data['city'] = encoder.fit_transform(input_data['city'].values.reshape(-1, 1))
    input_data['state'] = encoder.fit_transform(input_data['state'].values.reshape(-1, 1))
    input_data['postal_code'] = encoder.fit_transform(input_data['postal_code'].values.reshape(-1, 1))
    input_data['latitude'] = encoder.fit_transform(input_data['latitude'].values.reshape(-1, 1))
    input_data['longitude'] = encoder.fit_transform(input_data['longitude'].values.reshape(-1, 1))
    if len(input_data['attributes'].values.tolist()) != 0:
        attributes = [[val] for val in [', '.join([key for key in review if review[key] == 'True']) if review is not None else '' for review in input_data['attributes']]]
        input_data['attributes'] = encoder.fit_transform(attributes)
    input_data['categories'] = encoder.fit_transform(input_data['categories'].values.reshape(-1, 1))
    
    return input_data
```

```python
def query(user_id: str, user_query: str):
    assert user_id in user_ids

    user_reviews = review_data.loc[(review_data['user_id'] == user_id) & (review_data['business_id'].isin(business_data['business_id'].values.tolist()))]\
        .drop_duplicates(subset='business_id').sort_values(by='business_id').reset_index()
    if len(user_reviews.values.tolist()) < 5:
        print('Insufficient review count for prediction')
        return
    reviewed_businesses = business_data[business_data['business_id'].isin(user_reviews['business_id'].values.tolist())]\
        .drop_duplicates(subset='business_id').sort_values(by='business_id').reset_index()
    assert len(reviewed_businesses.values.tolist()) == len(user_reviews.values.tolist())
    
    output_data = user_reviews['stars']
    input_data = reviewed_businesses.drop(['business_id', 'hours'], axis=1).drop('index', axis=1)
    input_data = preprocessing(input_data)
    model = KNeighborsClassifier()
    model.fit(input_data.values, output_data.values)
    
    category = get_category(user_query)
    print("Category: " + category)
    businesses = business_data[business_data['categories'].str.contains(category, na=False)].drop(['business_id', 'hours'], axis=1).reset_index().drop('index', axis=1)
    business_names = businesses['name'].values.tolist()
    businesses = preprocessing(businesses)
    
    results = {}
    for i, business in businesses.iterrows():
        results[business_names[i]] = model.predict(business.values.reshape(1, -1))[0]
    recommendations = sorted(results, key=results.get, reverse=True)
    print('The best new places for you to check out are')
    for i in range(min(len(recommendations), 10)):
        print(str(i+1) + '. ' + recommendations[i] + " - " + str(results[recommendations[i]]) + " stars predicted")
    
query('AkBtT43dYcttxQ3qOzPBAg', 'driving theaters')
```
